```python
import numpy as np
```

```python
x = np.arange(1000) % 10
y = np.zeros(len(x))
```

```python
for n in range(1, len(x)):
    if x[n-1]==4 and x[n] == 5:
        y[n] = 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)
```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
yp = y.copy()
yp[len(x)//2:] = 0
```

```python
yp2 = np.random.randint(2, size=(len(x)))
```

```python
f1_score(yp2,y), accuracy_score(yp2, y)
```

```python
f1_scores = np.zeros(len(x))
for n in range(len(x)):
    yp = y.copy()
    yp[n:] = 0
    f1_scores[n] = f1_score(y,yp)
```

```python
plt.plot(np.arange(len(f1_scores))/len(f1_scores), f1_scores)
plt.title('f1 scores vs samples set to zero')
```
